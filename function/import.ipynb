# ML 스듀티오불러오기

```python
import requests
import json

url = "http://localhost:1234/v1/chat/completions"

headers = {
    "Content-Type": "application/json"
}

payload = {
    "model": "google/gemma-3-1b",
    "messages": [
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "너는 누구니?"}
    ],
    "temperature": 0.7
}

response = requests.post(url, headers=headers, data=json.dumps(payload))

print(response.status_code)
print(response.json()["choices"][0]["message"]["content"])
```

GeoPandas로 GeoJSON과 Shapefile(SHP)을 처리하는 기초 실무 가이드

1. 서론: GeoJSON · Shapefile · GeoPandas 개념 정리
1) Shapefile(SHP)

ESRI에서 만든 전통적인 벡터 공간데이터 포맷

실제로는 여러 파일의 묶음

.shp : 도형(geometry)

.shx : 인덱스

.dbf : 속성(attribute)

.prj : 좌표계(CRS)

장점: 오래된 표준, 대부분 GIS 소프트웨어에서 지원

단점: 파일 여러 개 관리 필요, 필드명 길이 제한

2) GeoJSON

JSON 기반의 웹 친화적 공간데이터 포맷

하나의 파일에 geometry + 속성 + 좌표계 포함

장점: 가볍고, 웹·API·파이썬 친화적

단점: 대용량일 경우 성능 저하

3) GeoPandas

Pandas + Shapely + Fiona + PyProj를 결합한 공간데이터용 DataFrame

핵심 객체: GeoDataFrame

geometry 컬럼에 Point / LineString / Polygon 저장
